### **Data Ingestion**
Nesta etapa realizamos a ingestão do dataset de transações financeiras utilizando Apache Spark.

**As principais atividades incluem:**
- leitura do arquivo CSV
- inferência automática de schema
- criação de uma temporary view para consultas utilizando Spark SQL
- inspeção inicial da estrutura dos dados

Este passo prepara os dados para as etapas seguintes de análise exploratória e engenharia de features.

```python
# Leitura e Validação
from pyspark.sql import SparkSession

caminho = "/Workspace/Users/henriquegmour4@gmail.com/Fraud Detection in Financial Transactions using PySpark/fraud-detection-spark/data/raw/credit_card_fraud.csv"

fraud_df = spark.read.csv(caminho, header=True, inferSchema=True, encoding="UTF-8")
fraud_df.createOrReplaceTempView("fraud_transactions")

# Validações
print("Data Ingestion")
print(f"Total de registros: {fraud_df.count()}")
print(f"Total de colunas: {len(fraud_df.columns)}")
print(f"Colunas: {fraud_df.columns}")
```

```python
# Verificação do Schema e Amostra
fraud_df.printSchema()
fraud_df.show(5)
```

```python
# Validações de Integridade
from pyspark.sql.functions import col, sum as spark_sum

# Verificar duplicatas
total = fraud_df.count()
unique = fraud_df.dropDuplicates().count()
print(f"Registros duplicados: {total - unique}")

# Verificar se coluna target existe
print(f"Coluna 'is_fraud' existe: {'is_fraud' in fraud_df.columns}")

# Contar valores nulos por coluna
fraud_df.select([
    spark_sum(col(c).isNull().cast("int")).alias(c)
    for c in fraud_df.columns
]).show()
```
